Lab01 part2 Lab task is to visualize the data set or the data using mat.lib


```python
import matplotlib.pyplot as plt

people=["Adithya","Sushruth","Sathwik","Gowtham"]
age =[27,10,5,1]
weight=[60,23,15,200]
height = [182,152,151,120]

plt.scatter(weight,height)
plt.title("Relationship between Weight and height")
plt.ylabel("Height in centimeters")
plt.xlabel("weight in kgs")
plt.show()

plt.scatter(age,weight)
plt.title("Relationship between Age and weight")
plt.ylabel("Weight in kgs")
plt.xlabel("Age")
plt.show()

```

Bar plot


```python
plt.figure(figsize=(12,4))
plt.title("peoples weight in kgs")
plt.bar(x=people,height=weight,width=0.6,alpha=0.6,color="black")

plt.xlabel("People",fontsize=15)
plt.ylabel("weight (in kgs)",fontsize=15)
plt.show()
```

Generate Histogram


```python
plt.figure(figsize=(7,5))

plt.hist(weight,color="red",edgecolor="k",alpha=0.5,bins=5)

plt.title("Histogram of patient weight")
plt.xlabel("Weight in kgs",fontsize=15)

plt.show()

```
